# 【章節三】單元六 回歸模型實作：多項式回歸 - Part A
* 資料描述(Data):二維資料點、線性分布
* 模型描述(Model):線性回歸模型(Linear Regression Model)

## 步驟一: 設定模型斜率(a)和截距(b)

```python
 a = 4
 b = 5
```

## 步驟二: 準備線性的二維資料點

```python
# 引入此步驟必要套件
import numpy as np
import random

x_positions = np.arange(-3,10,0.2)
y_position_list = []
y_position_correct_list = []

for x in x_positions:
  # y=ax+b
  y_position = a * x + b
  # 為每個 y 值賦與隨機性
  y_position_list.append(y_position + (random.random()-0.5)*10)
  # 基準線條
  y_position_correct_list.append(y_position)
```

```python
for y_random, y_correct in zip(y_position_list, y_position_correct_list):
  print(y_random, y_correct)
```

## 步驟三: 繪製資料的分布情況

```python
# 引入此步驟必要套件
import matplotlib.pyplot as plt

plt.figure(figsize=(8,6))
plt.plot(x_positions, y_position_list, "o")
plt.plot(x_positions, y_position_correct_list, label=f"y={a}x+{b}")
plt.xlabel("X-axis", fontsize=20)
plt.ylabel("Y-axis", fontsize=20)
plt.xticks(fontsize=16)
plt.yticks(fontsize=16)
plt.title("2D & Linear Data Points", fontsize=24)
plt.legend(fontsize=16)
plt.show()
```

## 步驟四: 進行模型擬合(Fitting)

```python
x_positions.reshape(-1,1)
```

```python
# 引入此步驟必要套件
from sklearn.linear_model import LinearRegression

x_positions_reshape = x_positions.reshape(-1,1)
y_positions_reshape = np.array(y_position_list).reshape(-1,1)

# 使用線性回歸演算法
LR_Algorithm = LinearRegression()
LR_Model = LR_Algorithm.fit(x_positions_reshape, y_positions_reshape)

LR_a = LR_Model.coef_[0][0]
LR_b = LR_Model.intercept_[0]

print("LR 斜率:",round(LR_a,2))
print("LR 截距:",round(LR_b,2))

LR_Y = []
for x in x_positions:
  LR_Y.append(LR_a * x + LR_b)
```

```python
x_new = 4.7
y_pred = LR_Model.predict(np.array(x_new).reshape(-1,1))[0][0]
y_pred
```

## 步驟五: 預測新資料點並視覺化

```python
x_new = 4.7

# 引入此步驟必要套件
import matplotlib.pyplot as plt

# 視覺化預測結果
plt.figure(figsize=(8,6))
plt.plot(x_positions, y_position_list, "o")
plt.plot(x_positions, LR_Y, label=f"y={round(LR_a,2)}x+{round(LR_b,2)}")

# 於下圖以紅點的形式呈現
plt.plot(x_new, y_pred, "o", markersize=10, c="red")

plt.xlabel("X-axis", fontsize=20)
plt.ylabel("Y-axis", fontsize=20)
plt.xticks(fontsize=16)
plt.yticks(fontsize=16)
plt.title("2D & Linear Data Points", fontsize=24)
plt.legend(fontsize=16)
plt.show()
```

# 【章節三】單元七 回歸模型實作：多項式回歸 - Part B
* 資料描述(Data):二維資料點、**非**線性分布
* 模型描述(Model):線性回歸模型、多項式回歸模型(Polynomial Regression Model)


## 步驟一: 設定模型參數 a,b,c

```python
a = 4  # 自行調整
b = 5  # 自行調整
c = -12  # 自行調整

# y = ax + bx**2 + c
```

## 步驟二: 準備非線性的二維資料點

```python
# 引入此步驟必要套件
import numpy as np
import random

x_positions = np.arange(-3, 10, 0.2)

y_position_list = []
y_position_correct_list = []
for x in x_positions:
  y_position = a * x + b * x**2 + c # y=ax+bx^2+c
  y_position_list.append(y_position + (random.random()-0.5) * 200) # 為每個 y 值賦予隨機性
  y_position_correct_list.append(y_position) # 基準線條
```

## 步驟三: 繪製資料的分布情況

```python
# 引入此步驟必要套件
import matplotlib.pyplot as plt

plt.figure(figsize=(8,6))
plt.plot(x_positions, y_position_list, "o")
plt.plot(x_positions, y_position_correct_list, label=f"y={str(a)}x+{str(b)}x^2+{str(c)}")
plt.title("2D & Unlinear Data Points", fontsize=24)
plt.xlabel("X-axis", fontsize=20)
plt.ylabel("Y-axis", fontsize=20)
plt.xticks(fontsize=16)
plt.yticks(fontsize=16)
plt.legend(fontsize=16)
plt.show()
```

## 步驟四: 進行線性模型擬合+視覺化

```python
# 引入此步驟必要套件
from sklearn.linear_model import LinearRegression

y_positions_reshape = np.array(y_position_list).reshape(-1,1)
x_positions_reshape = x_positions.reshape(-1,1)

# 使用線性回歸模型
LR_Model = LinearRegression()
LR_Model.fit(x_positions_reshape, y_positions_reshape)

print("[線性回歸模型的訓練結果]")
print(f"斜率(a)為 {round(LR_Model.coef_[0][0],2)}")
print(f"截距(b)為 {round(LR_Model.intercept_[0],2)}")

LR_a = LR_Model.coef_[0][0]
LR_b = LR_Model.intercept_[0]
LR_Y = []
for x in x_positions:
  LR_Y.append(LR_a * x + LR_b)

# 視覺化擬合結果
plt.figure(figsize=(8,6))
plt.plot(x_positions, y_position_list, "o")
plt.plot(x_positions, LR_Y, label=f"y={str(round(LR_a,2))}x+({str(round(LR_b,2))})")
plt.title("2D & Unlinear Data Points", fontsize=24)
plt.xlabel("X-axis", fontsize=20)
plt.ylabel("Y-axis", fontsize=20)
plt.xticks(fontsize=16)
plt.yticks(fontsize=16)
plt.legend(fontsize=16)
plt.show()
```

## 步驟五: 進行多項式模型擬合+視覺化

```python
# 引入此 Cell 會使用到的套件
from sklearn.linear_model import LinearRegression
# 引入讓 Polynomial Regression 可以 work 的方法
from sklearn.preprocessing import PolynomialFeatures

# 使用 PolynomialFeatures
poly = PolynomialFeatures(degree=2, include_bias=False)
poly_features = poly.fit_transform(x_positions.reshape(-1,1))
y_positions = np.array(y_position_list).reshape(-1,1)

# 使用多項式回歸模型
PR_Model = LinearRegression()
#PR_Model
PR_Model.fit(poly_features, y_positions)

print("[多項式回歸模型的訓練結果]")
print(f"a 為 {round(PR_Model.coef_[0][0],2)}; b 為 {round(PR_Model.coef_[0][1],2)}; c 為 {round(PR_Model.intercept_[0],2)}")
print(f"a 應接近{str(a)}; b 應接近於{str(b)}; c 應接近於{str(c)}")
```

```python
PR_Model.coef_, PR_Model.intercept_
```

```python
# 取得 PR_Model 各 coefficient
PR_a = PR_Model.coef_[0][0]
PR_b = PR_Model.coef_[0][1]
PR_c = PR_Model.intercept_[0]
PR_Y = []
for x in x_positions:
  PR_Y.append(PR_a * x + PR_b * x**2 + PR_c)
```

```python
PR_a,PR_b,PR_c
```

```python
# 視覺化擬合結果
plt.figure(figsize=(8,6))
plt.plot(x_positions, y_position_list, "o")
plt.plot(x_positions, PR_Y, label=f"y={(round(PR_a,2))}x+({(round(PR_b,2))})x^2+({(round(PR_c,2))})")
plt.title("2D & Unlinear Data Points", fontsize=24)
plt.xlabel("X-axis", fontsize=20)
plt.ylabel("Y-axis", fontsize=20)
plt.xticks(fontsize=16)
plt.yticks(fontsize=16)
plt.legend(fontsize=16)
plt.show()
```

## 步驟三: 辨識 數值型與類別型特徵

```python
numerical_attributes = []
categorical_attributes = []

for attr in Dataset.columns:
  attr_type = Dataset[attr].dtype
  if attr_type == "int64" or attr_type == "float64":
    numerical_attributes.append(attr)
  elif attr_type == "object":
    categorical_attributes.append(attr)

# 待會要針對 X 做 feautre scaling，所以先把 Y 拿掉
numerical_attributes.remove("price")

print(f"數值型的 Attributes 有: {numerical_attributes}")
print(f"類別型的 Attributes 有: {categorical_attributes}")
```

## 步驟四: 進行 One-Hot Encoding

```python
# 使用 pandas 的 get_dummies，可以迅速的對 Dataset 中屬於類別型的欄位進行 One-Hot Encoding
import pandas as pd

print(f"One-Hot Encoding 前有 {Dataset.shape[1]} 個欄位")

One_Hot_Dataset = pd.get_dummies(Dataset)

print(f"One-Hot Encoding 後有 {One_Hot_Dataset.shape[1]} 個欄位")
One_Hot_Dataset.head()
```

## 步驟五: 進行資料切分

```python
# 引入這部分會使用到的套件
from sklearn.model_selection import train_test_split

# 區分 X 與 Y
Y = One_Hot_Dataset["price"]
X = One_Hot_Dataset.drop(["price"],axis=1)

# 進行資料集切分
X_train, X_test, Y_train, Y_test = train_test_split(X,Y, test_size=0.3)
print(X_train.shape)
print(Y_train.shape)
print(X_test.shape)
print(Y_test.shape)
```
